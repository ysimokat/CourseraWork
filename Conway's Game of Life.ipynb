```python
import numpy as np
from numpy import random 
```

```python
def conway(s,p): #s is the size of the board and p is the population probability
    '''generates a board'''
    block = []
    for row in range(0,s):
        block.append([])
        for col in range(0,s):
            if p==0:
                block[row].append(random.choice([0]))
            elif p==1:
                block[row].append(random.choice([1]))
            else:
                block[row].append(random.randint(2))
                p += 0.1
    return block
```

```python
import matplotlib.pyplot as plt
def display(b1):
    '''display the board'''
    fig,ax = plt.subplots()
    ax.imshow(b1, cmap=plt.cm.summer, interpolation='nearest')
    ax.set_title("Conway's Game of Life")
    #ax.set_xlim([0,8])
    #ax.set_ylim([0,8])
    
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
 
    ax.spines['left'].set_position(('outward',10))
    ax.spines['bottom'].set_position(('outward',10))
    plt.show()
```

```python
def advance(b,t):
    ''' advance function plays game of life
        accepts as input a gameboard from conway function b and t time steps to take
        returns final state of board as a new board
    '''
    b1 = b
    for k in range(t):
        for x in range(len(b)):
            for y in range(len(b)):
                live = 0
                for i in range(x-1,x+2):
                    for j in range(y-1,y+2):
                        if i >= 0 and i < len(b) and j >= 0 and j < len(b):
                            if ([i,j]!=[x,y])==True and b[i][j] == 1:
                                live+=1
            if b[x][y] == 1:
                if live <2 or live >3: #under/over population
                    b1[x][y] = 0
                if live==2 or live==3: #next generation
                    b1[x][y]=1
            elif b[x][y] == 0:
                if live == 3:
                    b1[x][y]=1
                else:
                    b1[x][y]=0
        display(b1)
```
