10.	Create a labeled histogram of pop.

```python
plt.hist(pop, bins=8, alpha=0.8,color = 'orange')
plt.xlabel('pop')
plt.ylabel('Count')
plt.title('Histogram of POP') 
plt.grid(True) 
plt.show() 
```

11.	Create a labeled histogram of age.

```python
## Histogram for Age
plt.hist(age,bins=10,alpha=1,color='green')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Histogram of Age')
plt.grid(True)
plt.show()
```

12.	Create a labeled scatter plot of age (x-axis) and pop(y-axis).  Color the points in the scatterplot based on price.

```python
d = plt.scatter(age,pop,c=price,alpha = 0.5,marker='*')
plt.colorbar(d)
plt.xlabel('age')
plt.ylabel('pop')
plt.title('age v. pop')
plt.grid(True)
plt.show()
```

13.	Write a brief analysis of the scatterplot.

The plot shows that both the most expensive and cheapest houses are in the areas with smallest populations. Cheaper houses in smaller areas tend to be newer and older houses tend to be more expensive. The houses in the middle cost-wise are also in the middle age-wise, but are most abundant in areas with larger populations.
